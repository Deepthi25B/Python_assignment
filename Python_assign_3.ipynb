## While Loop, For loop and Function

Number Guessing Game

```python
import random

# Generate random number between 1 and 10
secret_number = random.randint(1, 10)

# Maximum number of attempts
attempts = 3

print("Welcome to the Number Guessing Game!")
print("You have", attempts, "attempts to guess the correct number.")

# While loop for guessing
while attempts > 0:
    guess = int(input("Guess the number (between 1 and 10): "))

    # Check if guess is out of range
    if guess < 1 or guess > 10:
        print("Your guess is out of range. Please guess a number between 1 and 10.")
        continue

    # Check if guess is correct
    if guess == secret_number:
        print("Congratulations! You guessed the correct number.")
        break

    elif guess > secret_number:
        print("Too high. Try again.")


    else:
        print("Too low. Try again.")

    # Reduce attempts after valid guess
    attempts -= 1
    print("Attempts left:", attempts)


else:
    print("Better luck next time! The correct number was", secret_number)
```

## For Loop:

Multiplication Table Generator

```python
number=int(input('Enter the number for which you want the multiplication table:'))
for i in range(1, 11):
    result = number * i
    print(number, "x", i, "=", result)
```

## Function:

BMI Calculator

```python
def calculate_bmi(weight, height):
    bmi = weight / (height ** 2)
    return bmi
weight = float(input("Enter your weight in kg: "))
height = float(input("Enter your height in meters: "))
bmi_result = calculate_bmi(weight, height)
print("Your BMI is:", round(bmi_result, 2))
```
