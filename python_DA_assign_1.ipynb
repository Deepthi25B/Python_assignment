## Panda series

```python
import pandas as pd

# Create Pandas Series with custom index
marks = pd.Series(
    [95, 92, 89, 85, 80],
    index=['Rank1', 'Rank2', 'Rank3', 'Rank4', 'Rank5']
)

print("Marks Series:")
print(marks)
```

```python
import pandas as pd

# Create the Series
marks = pd.Series(
    [95, 92, 89, 85, 80],
    index=['Rank1', 'Rank2', 'Rank3', 'Rank4', 'Rank5']
)

#  1. Access mark of 1st rank student using integer index position
print("1st Rank Mark (using integer index):", marks[0])

#  2. Retrieve marks of top 3 ranks using loc
print("Top 3 Ranks (using loc):")
print(marks.loc[['Rank1', 'Rank2', 'Rank3']])

#  3. Retrieve mark of 3rd rank student using iloc
print("3rd Rank Mark (using iloc):", marks.iloc[2])

#  4. Boolean mask for marks greater than 90
print("Ranks with marks greater than 90:")
print(marks[marks > 90])
```

```python
import pandas as pd

# Create the Series
marks = pd.Series(
    [95, 92, 89, 85, 80],
    index=['Rank1', 'Rank2', 'Rank3', 'Rank4', 'Rank5']
)

# 1. Modify the mark of the 1st rank student to 100
marks['Rank1'] = 100
print("After modifying Rank1 mark:")
print(marks)

# 2. Remove the entry corresponding to the last rank
marks = marks.drop('Rank5')
print("\nAfter removing last rank:")
print(marks)

# 3. Compute CGPA (marks divided by 10)
cgpa = marks / 10
print("\nCGPA:")
print(cgpa)
```

```python
import pandas as pd

# Create DataFrame
transactions = pd.DataFrame({
    'TransactionID': [101, 102, 103, 104, 105, 106, 107, 108, 109, 110],
    'ProductCategory': [
        'Electronics', 'Clothing', 'Electronics', 'Furniture',
        'Clothing', 'Electronics', 'Furniture', 'Clothing',
        'Furniture', 'Electronics'
    ],
    'Region': [
        'North', 'South', 'North', 'East', 'West',
        'North', 'East', 'West', 'South', 'North'
    ],
    'Amount': [200, 150, 300, 450, 200, 250, 300, 180, 350, 400]
})

# Display DataFrame
print("Full DataFrame:\n", transactions)

# Head and Tail
print("\nHead:\n", transactions.head())
print("\nTail:\n", transactions.tail())

# Shape
print("\nShape:", transactions.shape)

# Column Names
print("\nColumn Names:", transactions.columns)

# Data Types
print("\nData Types:\n", transactions.dtypes)

# Display only ProductCategory and Amount
print("\nSelected Columns:\n", transactions[['ProductCategory', 'Amount']])

# Retrieve last 3 columns using iloc
print("\nLast 3 Columns:\n", transactions.iloc[:, -3:])

# Filter Region = North and Amount > 200
filtered = transactions[(transactions['Region'] == 'North') & (transactions['Amount'] > 200)]
print("\nFiltered Rows:\n", filtered)

# Value counts of ProductCategory
print("\nProduct Category Counts:\n", transactions['ProductCategory'].value_counts())

# Unique Regions
print("\nUnique Regions:", transactions['Region'].unique())

# Group by Region and find mean Amount
print("\nMean Amount by Region:\n", transactions.groupby('Region')['Amount'].mean())
```

```python
import pandas as pd

# Create DataFrame
transactions = pd.DataFrame({
    'TransactionID': [101, 102, 103, 104, 105, 106, 107, 108, 109, 110],
    'ProductCategory': [
        'Electronics', 'Clothing', 'Electronics', 'Furniture',
        'Clothing', 'Electronics', 'Furniture', 'Clothing',
        'Furniture', 'Electronics'
    ],
    'Region': [
        'North', 'South', 'North', 'East', 'West',
        'North', 'East', 'West', 'South', 'North'
    ],
    'Amount': [200, 150, 300, 450, 200, 250, 300, 180, 350, 400]
})

# 1. Modify Amount for TransactionID 102 to 165
transactions.loc[transactions['TransactionID'] == 102, 'Amount'] = 165
print("After modifying Amount:\n", transactions)

# 2. Add Discount column (10% of Amount)
transactions['Discount'] = transactions['Amount'] * 0.10
print("\nAfter adding Discount column:\n", transactions)

# 3. Remove row with TransactionID 109
transactions = transactions[transactions['TransactionID'] != 109]
print("\nAfter removing TransactionID 109:\n", transactions)

# 4. Delete the Discount column
transactions = transactions.drop(columns=['Discount'])
print("\nAfter deleting Discount column:\n", transactions)
```
