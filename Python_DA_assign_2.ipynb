## Handling Missing Values

```python
import seaborn as sns
import pandas as pd
import numpy as np

# Load dataset
df = sns.load_dataset("taxis")

# 1. Check for missing values
print("Missing Values Count:\n")
print(df.isnull().sum())

# Identify columns with missing data
missing_columns = df.columns[df.isnull().any()]
print("\nColumns with Missing Values:", list(missing_columns))


# 2. Impute missing values

# Separate numerical and categorical columns
num_cols = df.select_dtypes(include=np.number).columns
cat_cols = df.select_dtypes(exclude=np.number).columns

# Impute numerical columns using median (robust to outliers)
for col in num_cols:
    if df[col].isnull().sum() > 0:
        df[col].fillna(df[col].median(), inplace=True)

# Impute categorical columns using mode
for col in cat_cols:
    if df[col].isnull().sum() > 0:
        df[col].fillna(df[col].mode()[0], inplace=True)

print("\nMissing Values After Imputation:\n")
print(df.isnull().sum())


# 3. Drop rows with missing values in critical columns (if any remain)

critical_columns = ['pickup', 'dropoff']  # Example critical columns
df = df.dropna(subset=critical_columns)

print("\nFinal Missing Values Check:\n")
print(df.isnull().sum())
```

## Visualizations using Matplotlib/Pandas Plot:

```python
#LINE CHART
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
df = sns.load_dataset("taxis")

# Convert pickup column to datetime
df['pickup'] = pd.to_datetime(df['pickup'])

# Sort by pickup time (important for line chart)
df = df.sort_values('pickup')

# Plot line chart
plt.figure(figsize=(10,5))
plt.plot(df['pickup'], df['fare'])

plt.title("Fare Over Time")
plt.xlabel("Pickup Time")
plt.ylabel("Fare")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
#BAR CHART
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
df = sns.load_dataset("taxis")

# Group by pickup_borough and sum fare
fare_by_borough = df.groupby('pickup_borough')['fare'].sum()

# Plot bar chart
plt.figure(figsize=(8,5))
fare_by_borough.plot(kind='bar')

plt.title("Total Fare by Pickup Borough")
plt.xlabel("Pickup Borough")
plt.ylabel("Total Fare")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# PIE CHART
import seaborn as sns
import matplotlib.pyplot as plt

# Load dataset
df = sns.load_dataset("taxis")

# Count trips by payment method
payment_counts = df['payment'].value_counts()

# Plot pie chart
plt.figure(figsize=(6,6))
plt.pie(payment_counts, labels=payment_counts.index, autopct='%1.1f%%')

plt.title("Distribution of Trips by Payment Method")
plt.tight_layout()
plt.show()
```

```python
# HISTOGRAM
import seaborn as sns
import matplotlib.pyplot as plt

# Load dataset
df = sns.load_dataset("taxis")

# Plot histogram for distance
plt.figure(figsize=(8,5))
plt.hist(df['distance'], bins=30)   # You can adjust bins (e.g., 20, 30, 40)

plt.title("Distribution of Trip Distance")
plt.xlabel("Distance")
plt.ylabel("Frequency")
plt.tight_layout()
plt.show()
```

```python
#BOX PLOT
import seaborn as sns
import matplotlib.pyplot as plt

# Load dataset
df = sns.load_dataset("taxis")

# Plot box plot
plt.figure(figsize=(8,5))
sns.boxplot(x='pickup_borough', y='tip', data=df)

plt.title("Tip Distribution by Pickup Borough")
plt.xlabel("Pickup Borough")
plt.ylabel("Tip Amount")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

### Visualization using Seaborn

```python
#COUNT PLOT
import seaborn as sns
import matplotlib.pyplot as plt

# Load dataset
df = sns.load_dataset("taxis")

# Create count plot
plt.figure(figsize=(8,5))
sns.countplot(x='pickup_borough', data=df)

plt.title("Number of Trips by Pickup Borough")
plt.xlabel("Pickup Borough")
plt.ylabel("Trip Count")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# SCATTER PLOT
import seaborn as sns
import matplotlib.pyplot as plt

# Load dataset
df = sns.load_dataset("taxis")

# Create scatter plot
plt.figure(figsize=(8,5))
sns.scatterplot(
    x='distance',
    y='fare',
    hue='pickup_borough',
    data=df
)

plt.title("Distance vs Fare by Pickup Borough")
plt.xlabel("Distance")
plt.ylabel("Fare")
plt.tight_layout()
plt.show()
```

```python
#HEATMAP
import seaborn as sns
import matplotlib.pyplot as plt

# Load dataset
df = sns.load_dataset("taxis")

# Select numerical columns
num_cols = ['distance', 'fare', 'tip', 'tolls', 'total']

# Create correlation matrix
corr_matrix = df[num_cols].corr()

# Plot heatmap
plt.figure(figsize=(8,6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f")

plt.title("Correlation Heatmap")
plt.tight_layout()
plt.show()
```

```python
# PAIR PLOT
import seaborn as sns
import matplotlib.pyplot as plt

# Load dataset
df = sns.load_dataset("taxis")

# Select required columns
selected_cols = ['distance', 'fare', 'tip', 'total', 'pickup_zone']

# Create pair plot
sns.pairplot(
    df[selected_cols],
    hue='pickup_zone'
)

plt.show()
```

```python
#Violin Plot
import seaborn as sns
import matplotlib.pyplot as plt

# Load dataset
df = sns.load_dataset("taxis")

# Create violin plot
plt.figure(figsize=(8,5))
sns.violinplot(
    x='payment',
    y='fare',
    data=df
)

plt.title("Fare Distribution by Payment Method")
plt.xlabel("Payment Method")
plt.ylabel("Fare")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
